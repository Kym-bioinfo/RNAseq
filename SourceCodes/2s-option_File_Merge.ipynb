＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝<br>
複数のcsvファイルを結合する<br>
pandasのデータフレームとしてcsv2つをそれぞれ読み込み、合わせて別のデータフレームに加算する<br>
読み込んだデータフレームをそれぞれdf1, df2、加算したものをdf3として<br>
df1+df2 =df3とし、df3をcsv出力した後にdf1=df3の処理を行うので、3つ以上のcsvを結合する場合は、df2のpathを書き換えて再度セルを実行すれば良い<br>
csvは毎回上書き出力するので、途中のファイルは残らない

```python
#@title df1のパスを入力
df1_path = "/content/drive/MyDrive/RNA-seq/Outputs/Hypoxia2/DC_only_tpm.csv" #@param {type:"string"}
df1 = pd.read_csv(df1_path)
print(df1)
```

```python
#@title df2のパスを入力
#@markdown df3を計算してcsv出力、df1の再定義までを行う
df2_path = "/content/drive/MyDrive/RNA-seq/Outputs/Hypoxia2/Stat_only_tpm.csv" #@param {type:"string"}
df2 = pd.read_csv(df1_path)
df3 = pd.merge(df1 , df2 , how = "inner")

print(df3)
print("存在している列名は次の通り")
print(list(df3))

df3.to_csv(exp_name + "contact.csv" , index = False)
df1 = df3
```

＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝<br>
結合したデータを並び替える<br>
まず、Geneidとproductをindex化し、残ったものを手動で並び替える

```python
#@title 2列をindex化、残った行名を表示
df3 = df3.set_index(["Geneid" , "product"])
print("存在している列名は次の通り")
print(list(df3))
```

並び替える順番に上に表示される列名を入力する<br>
クオーテーションマークは不要、スペースで区切ること<br>

```python
#@title 列の並び替え
#@markdown 必要な場合のみ実行すれば良い
index_list = "N_DC_1 N_DC_2 N_DC_3 H_DC_1 H_DC_2 H_DC_3" #@param {type:"string"}
index_list = index_list.split()
df4 = df3.reindex(index_list , axis=1)

print("\n行名ミスによるNaN値が無いことを確認")
df4
```

＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝＝<br>
データに平均行を追加する<br>
平均をとる開始行と終了行の行数を入力する<br>
また、平均列の列名を設定する

```python
#@title データの確認
print("並び替え後の順番はこの通り")
df5 = df4.copy(deep = True)
print(list(df5))
```

```python
#@title 平均を取る行数、平均列の列名の設定
#@markdown 平均を取る必要のある列の数だけ繰り返す<br>
#@markdown この際、列が追加されたことで列数が最初と異なるので注意<br>
#@markdown 間違えた場合は上のセルを実行することで元に戻すことができる
start = "1" #@param {type:"string"}
end = "3" #@param {type:"string"}
name = "N_DC_ave" #@param {type:"string"}

start = int(start) -1
end = int(end)
df5.insert(end , name , df5.iloc[: , start:end].mean(axis = 1).values)
df5.to_csv(exp_name + "avetpm.csv")
df5

```
